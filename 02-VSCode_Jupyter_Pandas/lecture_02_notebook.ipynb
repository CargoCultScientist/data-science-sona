## 2. Begriffe zur Orientierung

**Conda-Umgebung**  
Ein isolierter Arbeitsbereich mit eigener Python-Version und eigenen Paketen.

**Merksatz:**  
Interpreter und Kernel sollten möglichst aus derselben Umgebung stammen.

## 5. Vektoren und Matrizen

Im Kurs reicht zunächst folgende Intuition:

- Ein **Vektor** ist eine geordnete Reihe von Werten
- Eine **Matrix** ist eine rechteckige Anordnung von Werten in Zeilen und Spalten

Für einfache Beispiele verwenden wir `numpy`.


```python
import numpy as np

vector = np.array([2, 4, 6, 8])
matrix = np.array([[1, 2],
                   [3, 4]])

print("Vektor:")
print(vector)
print("Form:", vector.shape)

print("\nMatrix:")
print(matrix)
print("Form:", matrix.shape)
```

```python
print("Erstes Element des Vektors:", vector[0])
print("Element in Zeile 2, Spalte 1 der Matrix:", matrix[1, 0])
print("Mittelwert des Vektors:", vector.mean())
```

```python
print("Matrix + 1:")
print(matrix + 1)

print("\nMatrix mal 2:")
print(matrix * 2)
```

## 7. pandas und tabellarische Daten

`pandas` ist eine wichtige Bibliothek für tabellarische Daten.

Ein **DataFrame** ist eine Datenstruktur mit **Zeilen** und **Spalten**.  
Das ist eine gute Brücke zu Tabellen aus Excel oder SPSS.


```python
import pandas as pd

df = pd.DataFrame({
    "participant": [1, 2, 3, 4, 5],
    "stress": [2, 4, 5, 3, 4],
    "motivation": [4, 5, 3, 4, 2],
    "satisfaction": [3, 4, 2, 4, 3]
})

df
```

```python
print("Spaltennamen:", list(df.columns))
print("Anzahl Zeilen und Spalten:", df.shape)
print("Mittelwert Stress:", df["stress"].mean())
print("Mittelwert Motivation:", df["motivation"].mean())
```

```python
df["difference"] = df["motivation"] - df["stress"]
df
```
