## 4. Listen und Mittelwert

Hier verwenden wir eine Liste mit mehreren Werten und berechnen daraus einen Mittelwert.

```python
scores = [12, 18, 15, 20, 17]

print("Anzahl:", len(scores))
print("Summe:", sum(scores))
print("Mittelwert:", sum(scores) / len(scores))
```

## 6. Erster DataFrame

Ein `DataFrame` ist eine tabellarische Datenstruktur. Sie ist für viele Aufgaben in Data Science zentral.

```python
df = pd.DataFrame({
    "participant": [1, 2, 3, 4, 5],
    "motivation": [4, 5, 3, 4, 2],
    "stress": [2, 4, 5, 3, 4]
})

df
```

## 7. Einfache Auswertung

Jetzt berechnen wir die Mittelwerte zweier Spalten.

```python
print("Mittelwert Motivation:", df["motivation"].mean())
print("Mittelwert Stress:", df["stress"].mean())
```

## 8. Eine neue Spalte erzeugen

Wir berechnen hier eine einfache Differenz zwischen Motivation und Stress.

```python
df["difference"] = df["motivation"] - df["stress"]
df
```
